# Matplotlib

Matplotlib is a library created in 2003 for scientific data visualization.

Thanks to its architecture and systax, it allows us to do really nice plots and customize them for our purposes.

This library should be imported as follows:



```python
import matplotlib.pyplot as plt
```

## Let's start with some basic plots

We will first start our visualizations by making some plots of the following table:

| a | b |
|---|---|
| 2 | 3 |
| 4 | 7 |
| 5 | 5 |
| 6 | 9 |

When we are plotting with matplotlib we will need the data in a "vector shape", it can be a list, a tuple, a numpy array, among other (As you guys we will see in the following classes). So let's start by putting this into lists. 


```python
a = [2,4,5,6]
b = [3,7,5,9]
print(a,b)
```

Now lets actually plot it:

```python
plt.plot(a,b)
plt.show()
```

As you can see, make a plot in matplotlib is as easy as passing two lists with data to plt.plot(). However, we should consider some aspects such as the information contained in the plot, we will need title, name of the axis, legend in some cases and so on. Another consideration is how to display your plots:

There are 2 contexts:

1. Matplotlib is used in a terminal or scripts, `plt.show()` is a must.

2. Matplotlib is used in a IPython shell or a notebook, `plt.show()` is typically unnecessary.

Now let's make our plot better:

```python
plt.plot(a, b)
plt.xlabel("a values")
plt.ylabel("b values")
plt.title("My plot")
plt.show()
```

Now that we know the basics about plotting, lets do something more complicated. For this we are using numpy to generate the plots of various functions:

First, let's do the sin(x), then you will try cos(x):

Note: Remember to import numpy

```python
import numpy as np
x = np.linspace(1,10,20)
y = np.sin(x)

plt.plot(x , y)
plt.xlabel(r"x $m$")
plt.ylabel("f(x)")
plt.title(r"$sin(x)^2$ function") # r" $()$ "
```

```python
import numpy as np
x = np.linspace(1,10,100)
y = np.sin(x)
y2 = np.sin(x)*2

plt.plot(x , y)
plt.plot(x , y2)
plt.xlabel(r"x $m$")
plt.ylabel("f(x)")
plt.title(r"$sin(x)^2$ function") # r" $()$ "
```

We have now realized that we can plot any kind of function using matplotlib but when plotting various curves in the same plot we need to identify them. Next we will play with the legend and the styles of the line

### Legend

It allows to identify curves

### Line style

It allows for better identification of the curves




```python
plt.plot(x , y, label = r"$sin(x)$", color = "blue")
plt.plot(x , y2, label = r"$sin^2(x)$", color = "red", linestyle = "dashdot", marker = "<")
plt.xlabel(r"x")
plt.ylabel(r"$f(x)$")
plt.title(r"Functions")
plt.legend()
```

### Limits

We can set up the range of $x$ and $y$ axis



```python
## Limits

plt.plot(x , y, label = r"$sin(x)$", color = "blue")
plt.plot(x , y2, label = r"$sin^2(x)$", linestyle = "--", marker = "3", color = "#008000")
plt.xlabel(r"x")
plt.ylabel(r"$f(x)$")


plt.xlim(0,10)
plt.ylim(-1.5,1.5)



plt.title(r"Functions")
plt.legend()
```

### Position the legend

Sometimes the legend can hide a part of the plot that may be importat to us, for this reason we will learn how to locate the legend.

Some options for the location of the legend are:

'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'

```python
## Limits

plt.plot(x , y, label = r"$sin(x)$", color = "blue")
plt.plot(x , y2, label = r"$sin^2(x)$", linestyle = "--", marker = "3", color = "#008000")
plt.xlabel(r"x")
plt.ylabel(r"$f(x)$")


plt.xlim(0,10)
plt.ylim(-1.5,1.5)



plt.title(r"Functions")
plt.legend(loc = "right")
```

**bbox_to_anchor** gives more control for the placing of the legend. There is 4 possible inpluts for bbox_to_anchor -> (x0, y0, width, height)

```python
## Limits

plt.plot(x , y, label = r"$sin(x)$", color = "blue")
plt.plot(x , y2, label = r"$sin^2(x)$", linestyle = "--", marker = "3", color = "#008000")
plt.xlabel(r"x")
plt.ylabel(r"$f(x)$")


plt.xlim(0,10)
plt.ylim(-1.5,1.5)



plt.title(r"Functions")
plt.legend(bbox_to_anchor=(1, .05), loc = "lower left")
```

Finally, I want to show you that more things are custimizable in matplotlib, including the font type and size and the size of the image. We will the how to change the font type, the size type and the size of image now, but for some other customizations I recommend to google it. 

```python
plt.rcParams["font.family"] = "Ubuntu Mono"
plt.style.use('fivethirtyeight')
plt.rcParams['font.size'] = 15
plt.figure(figsize=(6,6))
for i in range(10):
    plt.plot(x, np.sin(x+i*0.4), label = f"fig{i}")

plt.title("Sin function plots")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend(bbox_to_anchor = (1.3,0.5), loc = "center right")

plt.grid(ls =":")


plt.show()
plt.close()
```

# Scatter plots

Scatter plots are useful when the dara oes not have a specific pattern, such as data from a probabilistic experiment. In this case, we want to see the points and maybe any kind of density of the plots. Lets see this situations by creating and potting random data.



```python
plt.style.use('default')
x_rand = np.random.rand(50)
y_rand = np.random.rand(50)

plt.scatter(x_rand, y_rand)
plt.title("Random data")
plt.xlabel("x")
plt.ylabel("y")
```

Alpha parameter

```python
x_rand = np.random.rand(10000)
y_rand = np.random.rand(10000)


plt.scatter(x_rand, y_rand, label = "triangles", alpha = 0.1)

plt.title("Random data")
plt.legend(bbox_to_anchor=(1.01,0.5))
plt.xlabel("x")
plt.ylabel("y")
```

```python
x_rand = np.random.random(50)
y_rand = np.random.random(50)
markers = [".", ",", "o", "v", "*", "+"]
count = 0
for marker in markers:

    plt.scatter(x_rand, y_rand + 2 * count, marker = marker, label = f"marker {marker}")
    count += 1
plt.legend(bbox_to_anchor = (1.01,0.5))

```

## Errorbars



```python
x = np.linspace(1,10,100)
error = 0.2*np.random.randn(100)
sin =  np.log(x)


plt.plot(x,sin)
plt.scatter(x, sin +error)
```

```python
plt.errorbar(x, sin, yerr = abs(error))
```

```python
plt.fill_between(x, sin - 0.2, sin + 0.2, alpha=0.2)
plt.plot(x,sin)
plt.scatter(x, sin +error)
```

## Anatomy of a figure

- Figure: It's the whole figure, we can thing of it as a conteiner of everything in the plot.
- Axes: Is the box than contain the region where we are going to plot.
- Axis: These objects set the scale and limits and generate ticks
- Title: Title of the figure
- xlabel, ylabel: Labels of the plot
- ticks: Small lines in the axis
